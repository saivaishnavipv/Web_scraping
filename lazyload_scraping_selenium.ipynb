```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import regex as re
import urllib
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
url="" ## enter your URL
```

```python
def scrollDown(browser, numberOfScrollDowns):
    body = browser.find_element_by_tag_name("body")
    while numberOfScrollDowns >=0:
        body.send_keys(Keys.PAGE_DOWN)
        numberOfScrollDowns -= 1
    return browser
```

```python
browser = webdriver.Chrome("D:\\chromedriver_win32\\chromedriver.exe")
browser.get(url)
browser = scrollDown(browser, 100)
all_hover_elements = browser.find_elements_by_class_name("post")
```

```python
# tags based on your requirement

img=[]
for hover_element in all_hover_elements:
    a_element = hover_element.find_element_by_tag_name("img")
    src = a_element.get_attribute("src")
    img.append(src)
```

```python
href=[]
for hover_element in all_hover_elements:
    a_element = hover_element.find_element_by_tag_name("a")
   
    src = a_element.get_attribute("href")
    href.append(src)
  
```

```python

text=[]            
for val in href:
    url =val
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    for val in soup.find_all('title'):
        text.append(val.text)
```

```python
filename=[]
for val in text:
    val=val.strip()
    val=val[:40]
    
    val=val.split(']')[0]+'.jpg'
    val=val.replace( "\'","_",)
    val=val.replace('/',"_")
    val=val.replace(' ','')
    val=val.replace('[','_')
    val=val.replace('"','')
    val=val.replace('>','_')
    val=val.replace('<','_')
    val=val.replace('=','_')
    val=val.replace('?','')
    val=val.replace('~','')

    filename.append(val) 
```

```python
list_dir="D:\\"

default_dir = os.path.join(list_dir,"Pictures")


opener = urllib.request.build_opener()
urllib.request.install_opener(opener)
i=0

for  url in img:
   
        print(i)
        url_new= url
        file = filename[i]
    
        img_data = opener.open(url_new)
        os.chdir(default_dir)
        f = open(file,"wb")
        f.write(img_data.read())
        f.close()
        i=i+1
```
